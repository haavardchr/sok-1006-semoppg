```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

energi_df = pd.read_csv('https://uit-sok-1006-v22.github.io/innleveringer/kraft-pris-prod.csv',delimiter=";")
print(energi_df)

energi_df["log_prod"] = np.log(energi_df["Produksjon i alt"])
energi_df["log_pris_oslo"] = np.log(energi_df["Oslo"])
print(energi_df)

#def log_fig
fig,ax=plt.subplots()
x = energi_df['log_prod']
y = energi_df['log_pris_oslo']
ax.set_ylabel('Pris Oslo')
ax.set_xlabel('Produksjon')
plt.scatter(x,y)

y=energi_df['log_pris_oslo']
pd.DataFrame(y)

x=pd.DataFrame(energi_df['log_prod'])
x['intercept']=1
x
```

```python
from statsmodels.regression.linear_model import OLS
res=OLS(y,x).fit()
print(res.summary())

x=np.linspace((min(energi_df['log_prod'])), max(energi_df['log_prod']), 5)
regression_line=res.params['intercept']+res.params['log_prod']*x
ax.plot(x, regression_line,color='red')
fig

new_df = energi_df.drop(21)
new_df

plt.figure()
x = new_df['log_prod']
y = new_df['log_pris_oslo']
fig,ax=plt.subplots()
plt.scatter(x,y)

y=new_df['log_pris_oslo']
pd.DataFrame(y)

x=pd.DataFrame(new_df['log_prod'])
x['intercept']=1

res=OLS(y,x).fit()

print(res.summary())

x=np.linspace((min(energi_df['log_prod'])), max(energi_df['log_prod']), 5)

regression_line=res.params['intercept']+res.params['log_prod']*x

ax.plot(x, regression_line,color='red')
fig
```
